```python
#Librerias a usar
import pandas as pd
import psycopg2
```

```python
# Conexión a la base de datos PostgreSQL
conexion = psycopg2.connect(
    host="localhost",
    port="5432",
    database="CIBERPOL",
    user="postgres",
    password="password"
)

# Leer el archivo Excel
df = pd.read_excel('delitos.xlsx')
df
```

```python
# Selecciona las columnas específicas a cargar
columnas_delito = ['Delito', 'Sección']
df = df[columnas_delito]
```

```python
# Iterar sobre las filas del DataFrame e insertarlas en la base de datos
cursor = conexion.cursor()
for indice, fila in df.iterrows():
    # Reemplazar los NaN por None
    fila = fila.where(pd.notnull(fila), None)
    # Construir la consulta SQL
    consulta = "INSERT INTO \"Delito\" (\"delito\", \"seccion\") VALUES (%s, %s)"
    # Ejecutar la consulta
    cursor.execute(consulta, fila.iloc[:])

# Confirmar la transacción y cerrar la conexión
conexion.commit()
cursor.close()
conexion.close()
```
